```python
import pandas as pd
q_df = pd.DataFrame([['male', 'teacher'], 
              ['male', 'engineer'],
              ['female', None],
              ['female', 'engineer']],columns=['Sex','Profession'])
q_df
```

```python
#缺失值填入字串'others'
q_df_fill = q_df.fillna('other')
qq=q_df_fill
qq
```

```python
#更進一步將字串做編碼。 此時用什麼方式做編碼比較適合?為什麼?

#無順序性編碼
pf = pd.get_dummies(q_df_fill[['Profession']])
q_df_fill = pd.concat([q_df_fill, pf],  axis=1)
print(q_df_fill)

#用第一個英文字母作排序值
Profession_mapping = {'teacher': 20, 'engineer': 5, 'other': 15}
qq['Profession'] = qq['Profession'].map(Profession_mapping)
qq
```
