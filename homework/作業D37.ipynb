## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
import sklearn
print(sklearn.__version__)
%matplotlib inline
```

## 讀入資料

```python
data_train = pd.read_csv("Titanic_train.csv")

missing_vals = data_train.isnull().sum()/len(data_train)
missing_vals.sort_values(ascending=False)
missing_vals = pd.DataFrame(missing_vals,columns=['missing_rate'])
print(missing_vals)
```

## 觀察 Age

```python
agepclass = sns.catplot(x="Pclass", y="Age", hue="Pclass",
               data=data_train,kind="box")
```

```python
agepclass = sns.catplot(x="Sex", y="Age", hue="Sex",
               data=data_train,kind="box")
```

```python
missing_index=data_train.isnull()
display(data_train[missing_index.values==True])
```

```python
ori_train=pd.DataFrame(data_train)
from sklearn import preprocessing
#下面兩行程式，讓 Male=1, FeMale-0.
le = preprocessing.LabelEncoder()
ori_train['Sex']=le.fit_transform(ori_train['Sex'])
train=ori_train.copy()
#del train['Survived','Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
del train['Survived']
del train['Name']
del train['SibSp']
del train['Parch']
del train['Ticket']
del train['Fare']
del train['Cabin']
del train['Embarked']
display(train)
```

```python
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(train)))
print('這個計算誤差太誇張了吧!')
```

```python
value_neighbors=3
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(train))
display(df_filled)
```

```python
imputer2 = KNNImputer(n_neighbors=value_neighbors,weights='distance')
df_filled2 = pd.DataFrame(imputer2.fit_transform(train))
display(df_filled2)
```
