題目:<br>
array1 = np.array([[10, 8], [3, 5]]) <br>
1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```python
import numpy as np
```

```python
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
array1 = np.array([[10, 8], [3, 5]])
invarray=np.linalg.inv(array1) 
print("反矩陣：",invarray)
multi=np.dot(array1,invarray)
print("矩陣相乘：",multi)
M=np.eye(2) #產生單位矩陣
print("檢查單位矩陣：", np.allclose(multi,M))
```

```python
#2. 運用上列array計算特徵值、特徵向量?
w, v = np.linalg.eig(array1)
print("特徵值：",w)
print("特徵向量：",v)
```

```python
#3. 運用上列array計算SVD?
print("SVD：",np.linalg.matrix_rank(array1))
```
