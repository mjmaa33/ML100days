### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
mu=0
sigma=2.5
x=np.linspace(-10, 10, 100)

```

```python
## Q1
#  提示: 用 cdf 來算
cumsum_probs =  stats.norm.cdf(1,mu, sigma)
print("小於1的機率=",cumsum_probs)
```
