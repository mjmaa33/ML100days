作業目標<br>
計算有缺失值的資料，統計量實作<br>
作業重點<br>
當遇到缺失值有函式可以處理，不須額外寫程式刪除<br>
計算統計量時不能出現缺失值

題目:<br>
english_score = np.array([55,89,76,65,48,70])<br>
math_score = np.array([60,85,60,68,np.nan,60])<br>
chinese_score = np.array([65,90,82,72,66,77])<br>
上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，今天第五位同學因某原因沒來考試，導致數學成績缺值，運用上列數據回答下列問題。<br>
1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
3. 用補考後資料找出與國文成績相關係數最高的學科?

```python
import numpy as np
```

```python
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,np.nan,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```python
#1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
englmean = np.nanmean(english_score)
englmax = np.nanmax(english_score)
englmin = np.nanmin(english_score)
englstd = np.nanstd(english_score)
mathmean = np.nanmean(math_score)
mathmax = np.nanmax(math_score)
mathmin = np.nanmin(math_score)
mathstd = np.nanstd(math_score)
chinesemean = np.nanmean(chinese_score)
chinesemax = np.nanmax(chinese_score)
chinesemin = np.nanmin(chinese_score)
chinesestd = np.nanstd(chinese_score)
print('english mean is',englmean)
print('english max is',englmax)
print('english min is',englmin)
print('english std is',englstd)
print('math mean is',mathmean)
print('math max is',mathmax)
print('math min is',mathmin)
print('math std is',mathstd)
print('chinese mean is',chinesemean)
print('chinese max is',chinesemax)
print('chinese min is',chinesemin)
print('chinese std is',chinesestd)
```

```python
#2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
math_score[4] = 55
mathmean1 = np.mean(math_score)
mathmax1 = np.max(math_score)
mathmin1 = np.min(math_score)
mathstd1 = np.std(math_score)
print('new math mean is',mathmean1)
print('new math max is',mathmax1)
print('new math min is',mathmin1)
print('new math std is',mathstd1)
```

```python
#3. 用補考後資料找出與國文成績相關係數最高的學科?
coefengchin = np.corrcoef(english_score,chinese_score)
coefmathchin = np.corrcoef(math_score,chinese_score)
print('english-chinese coef is',coefengchin)
print('math-chinese coef is',coefmathchin)
maxcoef = np.max([coefengchin[0,1],coefmathchin[0,1]])
print('與國文成績相關係數最高的學科是英文 corrcoef is ',maxcoef)
```
